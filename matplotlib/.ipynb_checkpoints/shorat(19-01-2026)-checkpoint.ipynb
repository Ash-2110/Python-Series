```python
import matplotlib.pyplot as plt

# 1. Data defining the subjects and marks
subjects = ['Maths', 'English', 'Science']
marks = [79, 35, 100]

# 2. Creating a bar graph
plt.bar(subjects, marks)

# 3. Adding labels to the axes
plt.xlabel('Subjects')
plt.ylabel('Marks')

# 4. Adding a title to the chart
plt.title('marks student')

# 5. Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data (marks of students)
marks = [45, 55, 60, 65, 70, 75, 80, 85, 90, 95, 60, 70, 80]

# Create histogram
plt.hist(marks, bins=5)

# Required labels and title
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.title("Marks Distribution")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [50, 55, 65, 70, 80]

# Create scatter plot
plt.scatter(x, y)

# Labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot Example")

# Display plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subjects = ['Maths', 'English', 'Science']
marks = [0.3, 0.2, 0.4]

# Create pie chart
plt.pie(marks, labels=subjects, autopct='%1.1f%%',startangle=90)

# Title
plt.title("Marks Distribution")

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subjects = ['Maths', 'English', 'Science', 'History', 'Geography', 'Physics', 'Chemistry', 'Biology']
marks = [95, 92, 94, 88, 85, 90, 89, 91]

# Explode (highlight Maths)
explode = [0.1, 0, 0, 0, 0, 0, 0, 0]

# Pie chart
plt.pie(
    marks,
    labels=subjects,
    autopct='%1.1f%%',
    startangle=90,
    explode=explode
)

# Title
plt.title("Marks Distribution")

# Display
plt.show()
```
