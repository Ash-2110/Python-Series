```python
dataa={
    "category":["electronics","furniture","electronics","furniture","office supplies"],
    "region":["east","west","south","north","west"],
    "sales":[1000,2000,1500,2000,3000],
    "profit":[100,200,150,200,300]

}

dt=pd.DataFrame(dataa)
```

```python
print(type(dt))
```

```python
dt
```

```python
catagory_sales=dt.groupby("category")["sales"].sum()
```

```python
import pandas as pd
```

```python
dfs=pd.DataFrame({
    "A":[1,2,None,4],
    "B":[5,None,4,6],
    "c":[9,4,5,9]
})
dfs
```

```python
print(dfs.isnull())
```

```python
print(dfs.isnull().sum())
```

```python
dfs['A'].fillna(dfs['A'].mean(),inplace=True)
```

```python
dfs['B'].fillna(dfs['B'].median(),inplace=True)
```

```python
dfs['c'].fillna(dfs['c'].mode()[0],inplace=True)
```
