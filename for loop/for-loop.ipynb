```python
#print with key and value by using for loop


word={
    "apple" : 2,
    "banana" : 3,
    "mango" : 4,
    "Cherry" : 5,
    "Litchi" : 6,
    "watermelon" : 7
}
for i in word:
    if i=="Litchi":
        break
    print(i,":",word[i])
```
